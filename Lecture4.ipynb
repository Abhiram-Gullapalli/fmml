```python
import numpy as np
import pandas as pd
```

```python
stocks = pd.Series(np.random.randn(1000))
```

```python
stocks = pd.DataFrame(np.random.randn(1000,4))
```

```python
stocks.index = pd.date_range('1/1/2020',periods=1000)
stocks.columns = ['a','b','c','d']
```

```python
stocks.cumsum().plot()
```

```python
stocks.to_excel('fmmlpractice.xlsx')
```
